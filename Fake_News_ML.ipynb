```python
# ---------------------- Single cell: Full pipeline with EXTENSIVE COMMENTS ----------------------
# Practical, direct, and heavily commented. Paste into one Google Colab cell and run.
# NOTE: If running in Colab, this will attempt to mount Google Drive (safe / optional).
# If your dataset is already loaded as df_true / df_fake in the environment, the loader will use them.

# ---------------- Imports & environment setup ----------------
import os
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

# sklearn components for vectorization, modeling, splitting, metrics
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, ConfusionMatrixDisplay, classification_report,
    roc_curve, auc, roc_auc_score
)

# NLTK for stopwords; download if absent.
import nltk
nltk.download('stopwords', quiet=True)
from nltk.corpus import stopwords

# Save/load objects
import joblib

# Deterministic seed — set once and reuse consistently for reproducibility
RANDOM_STATE = 42
np.random.seed(RANDOM_STATE)

# Configure visualization defaults (optional but tidy)
plt.style.use('seaborn-v0_8')
plt.rcParams.update({'figure.max_open_warning': 0})

# ---------------- Drive mount (Colab friendly) ----------------
# If you are in Colab, mounting Drive lets us load and persist files.
IN_COLAB = False
try:
    import google.colab
    IN_COLAB = True
except Exception:
    IN_COLAB = False

if IN_COLAB:
    from google.colab import drive
    # mount; if already mounted it is a no-op
    drive.mount('/content/drive', force_remount=False)
    print("Google Drive mounted at /content/drive")

# ---------------- Data loading (flexible, robust) ----------------
# Goals:
# - Prefer pre-existing df_true / df_fake (if you already prepared them)
# - Otherwise try common CSV filenames/paths
# - Validate text and label columns exist and are in expected format
# - Fail-fast with clear messages if dataset can't be found

try:
    # If variables exist in the workspace, we'll use them directly (useful during iterative runs)
    _ = df_true  # raise NameError if not defined
    _ = df_fake
    print("Using existing df_true and df_fake from notebook environment.")
except Exception:
    # Define candidate paths to look for CSV (you can add your path here)
    candidate_paths = [
        "/content/fake_new_ml_csv.csv",                    # Colab root
        "/content/drive/MyDrive/fake_new_ml_csv.csv",     # common Drive path
        "fake_new_ml_csv.csv"                             # local path fallback
    ]
    csv_path = None
    for p in candidate_paths:
        if os.path.exists(p):
            csv_path = p
            break

    if csv_path is None:
        raise FileNotFoundError(
            "No dataset found in expected locations. Upload 'fake_new_ml_csv.csv' to /content or Drive, "
            "or predefine df_true and df_fake in the environment."
        )

    # Load CSV robustly. Use engine='python' and skip bad lines to avoid parse errors on messy CSVs.
    print(f"Loading CSV from: {csv_path} (engine=python, on_bad_lines='skip')")
    df = pd.read_csv(csv_path, engine='python', on_bad_lines='skip')

    # Normalize column names (trim + lowercase) so we can detect 'Text' vs 'text' etc.
    df.columns = [c.strip().lower() for c in df.columns]

    # Identify label column (common names)
    label_col = None
    for candidate in ['label', 'class', 'target']:
        if candidate in df.columns:
            label_col = candidate
            break
    if label_col is None:
        raise ValueError("Could not find label column. Expected one of: 'label','class','target'.")

    # Identify text column
    text_col = None
    for candidate in ['text', 'content', 'article', 'body']:
        if candidate in df.columns:
            text_col = candidate
            break
    if text_col is None:
        raise ValueError("Could not find text column. Expected one of: 'text','content','article','body'.")

    # Keep only the essential columns to reduce memory (text + label)
    df = df[[text_col, label_col]].copy()

    # Coerce labels to numeric (some datasets use strings like 'FAKE'/'REAL')
    df[label_col] = pd.to_numeric(df[label_col], errors='coerce')
    # If coercion produced NaN (because labels were non-numeric), fallback to mapping strings
    if df[label_col].isna().any():
        # map unique string labels to integers (stable order)
        df[label_col] = df[label_col].astype(str)
        unique_labels = sorted(df[label_col].unique())
        mapping = {val: idx for idx, val in enumerate(unique_labels)}
        df[label_col] = df[label_col].map(mapping).astype(int)
        print("Non-numeric labels detected; mapped label values:", mapping)

    # Now split into df_true (label 0) and df_fake (label 1)
    # If labels are not strictly 0/1, we assume lowest unique label => 0, next => 1, etc.
    unique_labels = sorted(df[label_col].unique())
    if set(unique_labels) - {0, 1}:
        # map first unique -> 0, second -> 1 (this is a pragmatic binary mapping)
        mapper = {v: i for i, v in enumerate(unique_labels)}
        df[label_col] = df[label_col].map(mapper)
        print("Remapped labels to binary using:", mapper)

    # Filter and create consistent DataFrames
    df = df.dropna(subset=[text_col, label_col])  # drop empty rows
    # rename to consistent column name 'text'
    df = df.rename(columns={text_col: 'text', label_col: 'class_label'})
    df_true = df[df['class_label'] == 0].copy()
    df_fake = df[df['class_label'] == 1].copy()
    print("Loaded dataset. df_true shape:", df_true.shape, "df_fake shape:", df_fake.shape)

# ---------------- Preprocessing: cleaning function (single pass, efficient) ----------------
# Design choices:
# - Lowercase everything
# - Remove URLs (http... and www...)
# - Replace non-word characters with space (so tokens remain clean)
# - Remove stopwords (loaded once)
# - Do NOT aggressively stem/lemmatize to keep tokens interpretable for top-words analysis
stop_words = set(stopwords.words('english'))

def clean_text(text):
    """
    Clean text pipeline:
    1) Convert to string and lowercase
    2) Remove URLs
    3) Replace non-word chars with spaces
    4) Split and remove stopwords (fast list-based filter)
    5) Rejoin tokens into cleaned string (vectorizers expect strings)
    This is fast and robust for large corpora.
    """
    if not isinstance(text, str):
        text = str(text)
    text = text.lower()
    text = re.sub(r'http\S+|www\.\S+', ' ', text)            # remove URLs
    text = re.sub(r'[^\w\s]', ' ', text)                    # drop punctuation
    tokens = text.split()
    tokens = [tok for tok in tokens if tok not in stop_words and len(tok) > 1]
    return " ".join(tokens)

# Apply cleaning with a progress bar on both dataframes.
# Use .loc to avoid pandas SettingWithCopyWarning.
tqdm.pandas(desc="Cleaning df_true.text")
df_true.loc[:, 'text'] = df_true['text'].progress_apply(clean_text)

tqdm.pandas(desc="Cleaning df_fake.text")
df_fake.loc[:, 'text'] = df_fake['text'].progress_apply(clean_text)

# Drop any rows that became empty after cleaning
df_true = df_true[df_true['text'].str.strip() != ""].copy()
df_fake = df_fake[df_fake['text'].str.strip() != ""].copy()
print("After cleaning: df_true:", df_true.shape, "df_fake:", df_fake.shape)

# ---------------- Top-N words computation (safe & scalar outputs) ----------------
# We produce top-N (default 30) most frequent tokens for:
#   a) combined dataset (All)
#   b) true-only
#   c) fake-only
# The helper returns a list of (word, count) sorted descending.

from sklearn.feature_extraction.text import CountVectorizer

def get_top_n_words(corpus_series, n=30):
    """
    Returns a list of (word, count) pairs for the top n tokens in corpus_series.
    We use CountVectorizer with default tokenization because text is already cleaned.
    """
    vec = CountVectorizer()
    X = vec.fit_transform(corpus_series.astype(str))
    sums = np.array(X.sum(axis=0)).flatten()
    words = np.array(vec.get_feature_names_out())
    order = sums.argsort()[::-1]                     # indices of tokens sorted descending
    top_words = [(words[i], int(sums[i])) for i in order[:n]]
    return top_words

N = 30
df_all = pd.concat([df_true, df_fake], ignore_index=True)
words_all = get_top_n_words(df_all['text'], n=N)
words_true = get_top_n_words(df_true['text'], n=N)
words_fake = get_top_n_words(df_fake['text'], n=N)

# Build a DataFrame with scalar columns only (no tuples); easier to inspect and export
table_df = pd.DataFrame({
    'SrNo': list(range(1, N+1)),
    'word_both': [w for w,c in words_all],
    'count_both': [c for w,c in words_all],
    'word_true': [w for w,c in words_true],
    'count_true': [c for w,c in words_true],
    'word_false': [w for w,c in words_fake],
    'count_false': [c for w,c in words_fake],
})

print("Top words table created. First 8 rows preview:")
display(table_df.head(8))

# ---------------- Side-by-side visualization (3 panels horizontally) ----------------
# We use horizontal bars so long token labels don't overlap; largest value at the top.
fig, axes = plt.subplots(1, 3, figsize=(22, 8), sharex=False)

def horiz_bar(ax, words_counts, title, color):
    """
    Plot horizontal bar chart:
    - words_counts: list of (word, count) pairs
    - reverse the list so largest is at top (improves readability)
    """
    words = [w for w,c in words_counts][::-1]
    counts = [c for w,c in words_counts][::-1]
    y_positions = np.arange(len(words))
    ax.barh(y_positions, counts, color=color)
    ax.set_yticks(y_positions)
    ax.set_yticklabels(words, fontsize=9)
    ax.set_title(title, fontsize=12)
    ax.set_xlabel("Count")

horiz_bar(axes[0], words_all,  "Top 30 — All",  color='steelblue')
horiz_bar(axes[1], words_true, "Top 30 — True (label=0)", color='tab:green')
horiz_bar(axes[2], words_fake, "Top 30 — Fake (label=1)", color='tab:orange')

plt.suptitle("Top 30 Tokens: All vs True vs Fake — Side-by-side (horizontal bars)", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()

# ---------------- Optional: export the table (handy for manual inspection) ----------------
out_csv = "/content/top30_tokens_table.csv"
table_df.to_csv(out_csv, index=False)
print("Exported top-30 tokens table to:", out_csv)

# ---------------- ML pipeline: prepare train/test, vectorize, train, evaluate ----------------
# Key safety / correctness rules:
# 1) Always split into train/test BEFORE fitting vectorizer — no data leakage.
# 2) Use stratified split to preserve class balance.
# 3) Use sensible TF-IDF settings (max_features and ngram range) to control memory.
# 4) Evaluate multiple models and show standard metrics.

# Combine data into unified DataFrame for modeling
df_all_model = pd.concat([
    df_true.assign(class_label=0),
    df_fake.assign(class_label=1)
], ignore_index=True)
df_all_model = df_all_model.dropna(subset=['text', 'class_label']).reset_index(drop=True)

X = df_all_model['text'].values
y = df_all_model['class_label'].astype(int).values

# Stratified split (30% test)
X_train_text, X_test_text, y_train, y_test = train_test_split(
    X, y, test_size=0.30, stratify=y, random_state=RANDOM_STATE
)
print("Train / Test split sizes:", X_train_text.shape[0], "/", X_test_text.shape[0])

# Fit TF-IDF on train only — prevents leakage.
vectorizer = TfidfVectorizer(max_features=30000, ngram_range=(1,2), min_df=3)
X_train = vectorizer.fit_transform(X_train_text)
X_test  = vectorizer.transform(X_test_text)
print("TF-IDF vectorization complete. Train matrix shape:", X_train.shape)

# Define models to evaluate
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=RANDOM_STATE),
    'Decision Tree'      : DecisionTreeClassifier(random_state=RANDOM_STATE),
    'Random Forest'      : RandomForestClassifier(n_estimators=200, random_state=RANDOM_STATE, n_jobs=-1),
    'Naive Bayes'        : MultinomialNB()
}

# Train each model and capture metrics
results = {}
for name, model in models.items():
    print(f"\n--- Training {name} ---")
    model.fit(X_train, y_train)                # train
    preds = model.predict(X_test)              # discrete predictions
    # try to get probability scores for ROC if model supports it
    probs = None
    try:
        probs = model.predict_proba(X_test)[:,1]
    except Exception:
        probs = None

    # compute metrics
    acc = accuracy_score(y_test, preds)
    prec = precision_score(y_test, preds, zero_division=0)
    rec = recall_score(y_test, preds, zero_division=0)
    f1 = f1_score(y_test, preds, zero_division=0)
    roc_auc = roc_auc_score(y_test, probs) if (probs is not None) else None

    # store everything for later reporting/plotting
    results[name] = {
        'model_obj': model,
        'preds': preds,
        'probs': probs,
        'accuracy': acc,
        'precision': prec,
        'recall': rec,
        'f1': f1,
        'roc_auc': roc_auc,
        'confusion_matrix': confusion_matrix(y_test, preds)
    }

    print(f"{name} metrics -> Accuracy: {acc:.4f}, Precision: {prec:.4f}, Recall: {rec:.4f}, F1: {f1:.4f}, AUC: {roc_auc if roc_auc is not None else 'N/A'}")

# Summary table sorted by F1 (pragmatic selection criterion)
summary_df = pd.DataFrame([{
    'model': name,
    'accuracy': v['accuracy'],
    'precision': v['precision'],
    'recall': v['recall'],
    'f1': v['f1'],
    'roc_auc': (v['roc_auc'] if v['roc_auc'] is not None else np.nan)
} for name, v in results.items()]).sort_values(by='f1', ascending=False).reset_index(drop=True)

print("\nModel comparison (sorted by F1):")
display(summary_df)

# ---------------- Confusion matrices grid (2x2) ----------------
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.ravel()
for idx, (name, metrics) in enumerate(results.items()):
    cm = metrics['confusion_matrix']
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['True(0)', 'Fake(1)'])
    disp.plot(ax=axes[idx], values_format='d', cmap='Blues')
    axes[idx].set_title(name)
plt.tight_layout()
plt.show()

# ---------------- ROC curves for models that provide probabilities ----------------
plt.figure(figsize=(8,6))
plotted_any = False
for name, metrics in results.items():
    probs = metrics['probs']
    if probs is not None:
        fpr, tpr, _ = roc_curve(y_test, probs)
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f"{name} (AUC={roc_auc:.3f})")
        plotted_any = True

if plotted_any:
    plt.plot([0,1],[0,1],'--', color='grey')
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curves (models that support predict_proba)")
    plt.legend(loc='lower right')
    plt.show()
else:
    print("No models provided probability estimates; ROC curves not available.")

# ---------------- Detailed classification report for best model (by F1) ----------------
best_model_name = summary_df.loc[0, 'model']
print("Best model by F1:", best_model_name)
best_metrics = results[best_model_name]
best_model_obj = best_metrics['model_obj']
best_preds = best_metrics['preds']
print("\nClassification report for best model:")
print(classification_report(y_test, best_preds, digits=4))

# ---------------- Save best model and vectorizer (optional but recommended) ----------------
out_dir = "/content/models"
os.makedirs(out_dir, exist_ok=True)
best_model_path = os.path.join(out_dir, f"best_model_{best_model_name.replace(' ','_')}.joblib")
vectorizer_path = os.path.join(out_dir, "tfidf_vectorizer.joblib")

joblib.dump(best_model_obj, best_model_path)
joblib.dump(vectorizer, vectorizer_path)

print(f"Saved best model to: {best_model_path}")
print(f"Saved TF-IDF vectorizer to: {vectorizer_path}")
print("End of single-cell pipeline. If you want this split into multiple cells for stepwise debugging, say so.")
```
